```python
import numpy as np
import pandas as pd
df=pd.read_csv('lasso_sparse_150 (1).csv')
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_squared_error,r2_score
x=df[['x1','x2']]
y=df['y']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LassoCV(cv=5,random_state=42)
model.fit(x_train,y_train)  # fit the model in lasso
print(f'Best alpha:{model.alpha_}\n')
y_pred=model.predict(x_test)
coefficients=pd.Series(model.coef_, index=x.columns)  # Identify which features Lasso selects (non-zero coefficientts)
print('non-zero coefficients:')
print(f'{coefficients[coefficients != 0]}\n')
print(f'R^2:{r2_score(y_test,y_pred)}\n')
print(f'RMSE:{np.sqrt(mean_squared_error(y_test,y_pred))}')
```
